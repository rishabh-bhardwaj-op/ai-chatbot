SOLVEMATE - Your coding solution buddy 

1 : Import & Setup



```python
import os
from openai import OpenAI
```

2 : API Client Configuration


```python
openai = OpenAI(
    api_key="your grok api",
    base_url="https://api.groq.com/openai/v1"
)

print("Grok client (via OpenAI SDK) ready!")
```

3 : Conversation Data & Attempt Tracking 

```python
conversation_history = []
attempt_count = 0
```

4: Bot Function

```python
def bot(userinput):
    global attempt_count, conversation_history

    # Save user message
    conversation_history.append({"role": "user", "content": userinput})

    # Decide hint or full answer
    system_prompt = f"""
You are a friendly Python code tutor.  
Your rule:
1. First 2 replies should be only HINTS (no full solution).
2. and third reply, give full code + explanation.
3. DO NOT GIVE LENGHTY EXPLANATIONS
4: DO NOT REPLY TO ANY OTHER QUESTIONS

Current attempt count: {attempt_count}
"""
    messages = [{"role": "system", "content": system_prompt}] + conversation_history

    # API call
    response = openai.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=messages
    )

    bot_reply = response.choices[0].message.content

    # Save assistant reply
    conversation_history.append({"role": "assistant", "content": bot_reply})

    attempt_count += 1  

    return bot_reply
```

5 : Chat Loop

```python
print("SOLVEMATE is ready! (type 'exit' to stop)\n")

while True:
    user_text = input("You: ")

    if user_text.lower() == "exit":
        conversation_history = []
        attempt_count = 0
        print("Goodbye! All chat memory cleared.")
        break
        
    

    reply = bot(user_text)
    print("\nBot:", reply, "\n")
```
